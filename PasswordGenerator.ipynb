```python
import random
ls_small = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
ls_numeric = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
ls_symbol = ["~","`","!","@","#","$","%","^","&","*","(",")","_","-","+","=","{","[","}","]","|",":",";",",","<",">",".","?","/"]
ls_capital = []
valid = 0
num = 0
in_length = 0
             
for x in range(len(ls_small)):
    strcap = ls_small[x]
    strcap = strcap.upper()
    ls_capital.append(strcap)

in_type = input("What Type of Characters Should Password Contain\n 'CA' for Capital Alphabets\n 'CN' Capital Alphabets + Numbers\n 'CS' Capital Alphabets + Symbols\n 'SA' for Small Alphabets\n 'SN' Small Alphabets + Numbers\n 'SS' Small Alphabets + Symbols\n 'MA' for Capital + Small Alphabets\n 'AN' for Small + Capital Alphabets + Numbers\n 'AS' for Small + Capital Alphabets + Symbols\n 'N' For Numbers\n 'NS' for Numbers + Symbols\n 'S' For Symbols\n 'L' For All\n\t Your Input: ").upper()
while valid == 0:
    if in_type != "CA" and in_type != "CN" and in_type != "CS" and in_type != "SA" and in_type != "SN" and in_type != "SS" and in_type != "MA" and in_type != "AN" and in_type != "AS" and in_type != "N" and in_type != "NS" and in_type != "S" and in_type != "L":
        print("Invalid Input")
        in_type = input("What Type of Characters Should Password Contain\n 'CA' for Capital Alphabets\n 'CN' Capital Alphabets + Numbers\n 'CS' Capital Alphabets + Symbols\n 'SA' for Small Alphabets\n 'SN' Small Alphabets + Numbers\n 'SS' Small Alphabets + Symbols\n 'MA' for Capital + Small Alphabets\n 'AN' for Small + Capital Alphabets + Numbers\n 'AS' for Small + Capital Alphabets + Symbols\n 'N' For Numbers\n 'NS' for Numbers + Symbols\n 'S' For Symbols\n 'L' For All\n\t Your Input: ").upper()
    else:
        valid = 1
while valid == 1:
    try:
        in_length = int(input("Enter length of Password: "))
        print("")
    except ValueError:
        print("\nInvalid Input Please Enter Numbers Only")
    if in_length > 0:
        valid = 0
    else:
        valid = 1
        
if in_type == "CA":
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        print(random.choice(ls_capital), end='')
elif in_type == "CN":
    ls_CN = [ls_capital, ls_numeric]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_CN))
        print(random.choice(ls_CN[num]), end='')
elif in_type == "CS":
    ls_CS = [ls_capital, ls_symbol]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_CS))
        print(random.choice(ls_CS[num]), end='')
        
elif in_type == "SA":
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        print(random.choice(ls_small), end='')
elif in_type == "SN":
    ls_SN = [ls_small, ls_numeric]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_SN))
        print(random.choice(ls_SN[num]), end='')
elif in_type == "SS":
    ls_SS = [ls_small, ls_symbol ]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_SS))
        print(random.choice(ls_SS[num]), end='')
        
elif in_type == "MA":
    ls_MA = [ls_small, ls_capital]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_MA))
        print(random.choice(ls_MA[num]), end='')
elif in_type == "AN":
    ls_AN = [ls_small, ls_capital, ls_numeric]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_AN))
        print(random.choice(ls_AN[num]), end='')
elif in_type == "AS":
    ls_AS = [ls_small, ls_capital, ls_symbol]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_AS))
        print(random.choice(ls_AS[num]), end='')
        
elif in_type == "N":
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        print(random.choice(ls_numeric), end='')
elif in_type == "NS":
    ls_NS = [ls_numeric, ls_symbol]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_NS))
        print(random.choice(ls_NS[num]), end='')
elif in_type == "S":
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        print(random.choice(ls_symbol), end='')
elif in_type == "L":
    ls_ALL = [ls_small, ls_capital, ls_numeric, ls_symbol]
    print("Password Generated: \n====================================================================")
    for x in range(in_length):
        num = random.randrange(len(ls_ALL))
        print(random.choice(ls_ALL[num]), end='')
else:
    print("Wrong Input")
print("\n====================================================================")
```
