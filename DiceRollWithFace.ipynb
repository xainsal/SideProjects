```python
def diceFace():
    import random
    x = random.randrange(1,7)
    return x

dice = ""

while dice == "":
    dice = input("Enter '0' to Exit\nPress Enter To Roll The Dice: ")
    print("\n")
    dice_Val = diceFace()
    if dice_Val == 1:
        print("---------\n|       |\n|   o   |\n|       |\n---------\n")
    elif dice_Val == 2:
        print("---------\n| o     |\n|       |\n|     o |\n---------\n")
    elif dice_Val == 3:
        print("---------\n| o     |\n|   o   |\n|     o |\n---------\n")
    elif dice_Val == 4:
        print("---------\n| o   o |\n|       |\n| o   o |\n---------\n")
    elif dice_Val == 5:
        print("---------\n| o   o |\n|   o   |\n| o   o |\n---------\n")
    elif dice_Val == 6:
        print("---------\n| o   o |\n| o   o |\n| o   o |\n---------\n")
    elif dice == 0:
        break
```
